```python

contain_tables = soup.select('#content > div.rst_sch tbody')
tables=[]
for index,t in enumerate(contain_tables):
    table_list=[]
    rows=t.select('tr')

    for row in rows:
        row_list = []
        rank = row.select_one('td:nth-child(1)')
        title = row.select_one('td.tal > span.ellip.per90 > a')
        date = row.select_one('td:nth-child(3)')
        money = row.select_one('td:nth-child(4)')
        moneyr= row.select_one('td:nth-child(5)')
        moneyinc = row.select_one('td:nth-child(6)')
        moneyacc = row.select_one('td:nth-child(7)')
        watcher =  row.select_one('td:nth-child(8)')
        watcherinc = row.select_one('td:nth-child(9)')
        accwatcher = row.select_one('td:nth-child(10)')
        screen = row.select_one('td:nth-child(11)')
        times = row.select_one('td:nth-child(12)')

        rank = rank.get_text().strip()
        title = title.get_text().strip()
        date = date.get_text().strip()
        money = money.get_text().strip()
        moneyr = moneyr.get_text().strip()
        moneyinc = moneyinc.get_text().strip()
        moneyinc = moneyinc.replace('\t', '')
        moneyinc = moneyinc.replace('\n', '')
        moneyacc = moneyacc.get_text().strip()
        watcher = watcher.get_text().strip()
        watcherinc = watcherinc.get_text().strip()
        watcherinc = watcherinc.replace('\t', '')
        watcherinc = watcherinc.replace('\n', '')
        accwatcher = accwatcher.get_text().strip()
        screen = screen.get_text().strip()
        times = times.get_text().strip()

        row_list.append(rank)
        row_list.append(title)
        row_list.append(date)
        row_list.append(money)
        row_list.append(moneyr)
        row_list.append(moneyinc)
        row_list.append(moneyacc)
        row_list.append(watcher)
        row_list.append(watcherinc)
        row_list.append(accwatcher)
        row_list.append(screen)
        row_list.append(times)

        table_list.append(row_list)
    
    tables.append(table_list)
    if index ==6:
        break
print(tables)
print(len(tables))

```
