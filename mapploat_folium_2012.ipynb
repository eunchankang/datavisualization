```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ptl
import folium
import webbrowser
import pickle as pk
```

```python
with open ('t.pkl','rb') as f:
    t=pk.load(f)
f.close()

df = pd.DataFrame(t)
df.head()
```

```python
df2=df[df.year>=2012]
df3=df[df.year<2012]
```

```python
dfa12=df2[['country','latitude','longitude']]
dfb12=df3[['country','latitude','longitude']]
```

```python
dfa12=dfa12.dropna()
dfb12=dfb12.dropna()
```

```python
dfw = df[['country','latitude','longitude']]
dfw=dfw.dropna()
```

```python
latitude = -14.6048
longitude = -59.0625
m = folium.Map(location=[latitude, longitude], zoom_start=2)

for lat,lon in zip(dfw['latitude'],dfw['longitude']):
     folium.CircleMarker([lat, lon],
                            
                            radius=0.001,
                            
                            fill=True,
                            fill_opacity=0.7,
                            fill_color='r',
                           ).add_to(m)
m.save('worldplot.html')
```

```python
m1 = folium.Map(location=[latitude, longitude], zoom_start=2)

for lat,lon in zip(dfb12['latitude'],dfb12['longitude']):
     folium.CircleMarker([lat, lon],
                            
                            radius=0.001,
                            
                            fill=True,
                            fill_opacity=0.7,
                            fill_color='r',
                           ).add_to(m1)
m1.save('before2012.html')
```

```python
m2 = folium.Map(location=[latitude, longitude], zoom_start=2)
for lat,lon in zip(dfa12['latitude'],dfa12['longitude']):
     folium.CircleMarker([lat, lon],
                            
                            radius=0.001,
                            
                            fill=True,
                            fill_opacity=0.7,
                            fill_color='r',
                           ).add_to(m2)
m2.save('after2012.html')



```
