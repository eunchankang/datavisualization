```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as ptl
import pickle as pk




```

```python
with open('t.pkl','rb') as f:
    t = pk.load(f)
```

```python
df = pd.DataFrame(t)
#df.head()
```

```python
df2=df[df.year>=2012]
df3=df[df.year<2012]
```

```python

df_c = df[['latitude','longitude','casualities']]
#0 값을 nan으로 변환
df_c['casualities']=df_c['casualities'].apply(lambda x: np.nan if x==0 else x)
#nan 제외
df_c =df_c.dropna()
```

```python
df_k = df[['latitude','longitude','Killed']]
#0 값을 nan으로 변환
df_k['Killed']=df_k['Killed'].apply(lambda x: np.nan if x==0 else x)
#nan 제외
df_k = df_k.dropna()
```

```python
BBox=(-173,185.2,-73.5,85.3)

mp1= ptl.imread('map2.png')
fig, ax = ptl.subplots(figsize=(15,15))
ax.scatter(df_c.longitude, df_c.latitude, zorder = 1, c='k',s=0.1 *df_c.casualities,edgecolors="grey",alpha =0.6 )
ax.scatter(df_k.longitude,df_k.latitude, zorder =1, c ='pink', s = 0.1*df_k.Killed, edgecolors="red", linewidth = 0.3 )

ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(mp1,zorder =0, extent = BBox, aspect = 'equal')
```
